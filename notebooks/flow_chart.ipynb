```python
from flowgiston import FlowgistonChart, flowgiston_base
Base = flowgiston_base()
class Structure(Base):

    shape = 'diamond'
```

```python
# dir(chart)
```

```python
chart = FlowgistonChart(Base)
start = chart.start('run_logs_pipeline(args)')
step1 = start.edge(chart.process('get: \n -data_folder \n -parent_path'))
step2 = step1.edge(chart.process('logs_list_names = get_all_logs_from_path(parent_path, data_folder)'))
step3 = step2.edge(chart.Structure.node('for i in logs_list_names:'))
step4 = step3.yes(chart.process('get_txt_from_logs(i, parent_path, data_folder)'))
step5 = step3.no(chart.process('print(File cannot be extracted))'))
step6 = step4.edge(chart.process('convert_txts_to_csv()'))
```
